```python
numeroLimite = int(input("Escreva um número limite: "))

def SomaImpares(numeroLimite):
    somaTotal = 0
    for x in range(numeroLimite + 1):
        if(x % 2 == 0):
            continue
        else:
            somaTotal += x
    print(f"A soma de todos os número ímpares até {numeroLimite} é: {somaTotal}")

SomaImpares(numeroLimite)
```
